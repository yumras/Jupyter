# 4. scikit-learn 예제

### 4-1. 꽃잎 기준 표시

```python
# PANDAS 패키지 불러오기
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# CSV 파일을 읽어 Data Frame 변수에 저장하기
flowerData = pd.read_csv("../dataset/IRIS.csv")

name = flowerData[['SPECIES']]
# feautre = 클러스터의 기준 = (꽃받침 길이 넓이)
feature = flowerData[['PETALLENGTH','PETALWIDTH']]

# create model and prediction
model = KMeans(n_clusters=3, algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['PREDICT']

# concatenate labels to df as a new column
r = pd.concat([name,feature,predict],axis=1)

# scatter plot
plt.scatter(r['PETALLENGTH'],r['PETALWIDTH'],c=r['PREDICT'],alpha=0.5)
# plt.scatter(center_x,center_y,s=50,marker='D', c='r')
plt.show()

r.to_csv("d:/irisresult.csv", index=False, encoding="utf-8-sig")
```

### 4-2. 꽃받침 면적, 꽃잎 면적 기준

```python
# PANDAS 패키지 불러오기
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# CSV 파일을 읽어 Data Frame 변수에 저장하기
flowerData = pd.read_csv("../dataset/IRIS.csv")

name = flowerData[['SPECIES']]
# feautre = 클러스터의 기준 = (꽃잎 길이 넓이, 꽃받침 길이 넓이)
feature = flowerData[['SEPALLENGTH','SEPALWIDTH','PETALLENGTH','PETALWIDTH']]

# create model and prediction
model = KMeans(n_clusters=3, algorithm='auto')
model.fit(feature)
predict = pd.DataFrame(model.predict(feature))
predict.columns=['PREDICT']

# concatenate labels to df as a new column
r = pd.concat([name,feature,predict],axis=1)

# scatter plot
plt.scatter((r['SEPALLENGTH']*r['SEPALWIDTH']),(r['PETALLENGTH']*r['PETALWIDTH']),c=r['PREDICT'],alpha=0.5)
# plt.scatter(center_x,center_y,s=50,marker='D', c='r')
plt.show()

r.to_csv("d:/irisresult_2.csv", index=False, encoding="utf-8-sig")
```
