### 2. 원하는 열(컬럼) 조회하기

```python
# PANDAS 패키지 불러오기
import pandas as pd
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_csv("../dataset/customerdata.csv")

# 컬럼해더 재정의
customerData.columns = ['CUSTID','AVGPRICE','EMI','DEVICECOUNT','PRODUCTAGE','CUSTTYPE']

# 데이터프레임 생성
customerDataDf = pd.DataFrame(customerData)

# 데이터프레임 리스트 활용하기
customerDataC = customerDataDf[['CUSTID','EMI']]
customerDataC.head()
```

### 3. 인덱스 활용하여 조회하기

#### 데이터프레임명.iloc[행시작 : 행끝 , 열시작 : 열끝 ] or [ 행시작 : 행끝 , [열 1, 열 3]]
#### # 앞에는 행 인덱스 뒤에는 열 인덱스   ‘:’ 만 활용 시 전체

```python
# PANDAS 패키지 불러오기
import pandas as pd
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_csv("../dataset/customerdata.csv")

# 컬럼해더 재정의
customerData.columns = ['CUSTID','AVGPRICE','EMI','DEVICECOUNT','PRODUCTAGE','CUSTTYPE']

# 데이터프레임 생성
customerDataDf = pd.DataFrame(customerData)

# 인덱스로 조회하기
dfc1 = customerData.iloc[0:1,:]
dfc1
```

### 3. 인덱스 활용하여 조회하기 2 (컬럼명)

#### 데이터프레임명.loc[행시작 : 행끝 , 열시작 : 열끝 ]  or [ 행시작 : 행끝 , ['열 1 컬럼명', '열 3 컬럼명']]
#### # 앞에는 행 인덱스 뒤에는 열 인덱스   ‘:’ 만 활용 시 전체

```python
# PANDAS 패키지 불러오기
import pandas as pd
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_csv("../dataset/customerdata.csv")

# 컬럼해더 재정의
customerData.columns = ['CUSTID','AVGPRICE','EMI','DEVICECOUNT','PRODUCTAGE','CUSTTYPE']

# 데이터프레임 생성
customerDataDf = pd.DataFrame(customerData)

# 인덱스로 조회하기
dfc2 = customerData.loc[0:2, ["CUSTID","AVGPRICE"]]
dfc2
```
